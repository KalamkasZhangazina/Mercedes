```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

data = pd.read_csv("mercedes.csv")
print(data.isnull().sum())
```

```python
sns.set_style("darkgrid")
sns.countplot(y = "year", data = data, palette = "colorblind")
plt.xlabel("year")
plt.ylabel("price")
plt.show()
```

```python
very_new_car = len(data[data["year"] >= "2015 г."])
new_car = len(data[data["year"] >= "2000 г."])
old_car = len(data) - very_new_car - new_car
print("\n Очень новых машин: ", very_new_car, "\n", "Новых машин: ", new_car, "\n", "Старых машин: ", old_car)
```
